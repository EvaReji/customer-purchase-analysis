# Day 5 – RFM Customer Analysis

Goal : The aim of this step is to study customer purchasing behavior using RFM analysis. In this method, we look at three things:

- **Recency** – how recently a customer made a purchase  
- **Frequency** – how often a customer buys  
- **Monetary** – how much money a customer spends  

This helps understand which customers are more valuable, which customers buy frequently, and which customers may not have purchased recently.

### STEP 1 – Create Snapshot Date

To calculate **Recency**, we need a reference date.  
This is usually the day after the last transaction date in the dataset.

We find the latest `InvoiceDate` and add one day to create the snapshot date.

```python
snapshot_date = df['InvoiceDate'].max() + pd.Timedelta(days=1)
snapshot_date
```

### STEP 2 – Build RFM Table

In this step, I grouped the dataset by **CustomerID** and calculated the Recency, Frequency, and Monetary values for each customer.  
This creates an RFM table that summarizes customer purchasing behavior based on their transaction history.

```python
# Calculate Recency
recency = df.groupby('CustomerID')['InvoiceDate'].max()
recency = (snapshot_date - recency).dt.days

# Calculate Frequency
frequency = df.groupby('CustomerID')['InvoiceNo'].nunique()

# Calculate Monetary
monetary = df.groupby('CustomerID')['TotalPrice'].sum()

# Combine into RFM table
rfm = pd.concat([recency, frequency, monetary], axis=1)

rfm.head()
```

### STEP 3 – Rename Columns

To make the RFM table easier to understand, I renamed the columns to **Recency**, **Frequency**, and **Monetary**.

```python
rfm.columns = ['Recency', 'Frequency', 'Monetary']
rfm.head()
```

To better understand customer behavior, I checked the summary statistics of the RFM table.  
This helps see the overall distribution of Recency, Frequency, and Monetary values.

# Day 6 – Customer Segmentation using RFM Analysis

Goal : Segment customers based on **Recency, Frequency, and Monetary (RFM) values** to better understand customer purchasing behavior.

By assigning RFM scores and grouping customers into segments, we can identify:

- High-value customers
- Loyal customers
- Customers who may become inactive

This analysis helps businesses focus on **customer retention and revenue growth**.

### STEP 1 – Verify the RFM Table

Before performing customer segmentation, the RFM table is checked to ensure it contains the required metrics: **Recency, Frequency, and Monetary**.

```python
rfm.head()
```

```python
rfm.shape
```

### Observation
The RFM table contains **4,338 customers**, confirming the dataset is ready for further analysis.

### STEP 2 – Assign RFM Scores

To prepare for customer segmentation, **RFM scores** are assigned to Recency, Frequency, and Monetary values.  
The data is divided into **quartiles**, where each customer receives a score from **1 to 4** based on their position in the distribution.

Customers with **recent purchases, higher purchase frequency, and higher spending** receive higher scores.


```python
# Recency: lower recency → higher score
rfm['R_score'] = pd.qcut(rfm['Recency'], 4, labels=[4,3,2,1])

# Frequency: higher frequency → higher score
rfm['F_score'] = pd.qcut(rfm['Frequency'].rank(method='first'), 4, labels=[1,2,3,4])

# Monetary: higher spending → higher score
rfm['M_score'] = pd.qcut(rfm['Monetary'], 4, labels=[1,2,3,4])

rfm.head()
```

### STEP 3 – Create RFM Score 

Combine the **R, F, M scores** into a single 3-digit score for each customer.  
This makes it easier to **segment customers** in the next step.

```python
# Combine R, F, M scores into a single string
rfm['RFM_Score'] = rfm['R_score'].astype(str) + rfm['F_score'].astype(str) + rfm['M_score'].astype(str)

# Check the result
rfm[['R_score','F_score','M_score','RFM_Score']].head()
```

### STEP 4 – Customer Segmentation 

In this step, we categorize customers into simple segments based on their **R_score** and **F_score**.  
This helps identify **Top Customers, Loyal buyers, and those needing engagement**.

### Segment Rules

| Segment         | Meaning |
|-----------------|---------|
| Top Customer    | Most recent & frequent buyers |
| Loyal           | Frequent buyers |
| Potential       | Recent buyers, not frequent yet |
| At Risk         | Not very recent buyers |
| Lost            | Inactive or low-value customers |

```python
#Customer Segmentation using DataFrame.loc

# Assign default segment
rfm['Segment'] = 'Lost'

# Top Customer: Recent and frequent buyers
rfm.loc[(rfm['R_score'] == 4) & (rfm['F_score'] >= 3), 'Segment'] = 'Top Customer'

# Loyal: Frequent buyers
rfm.loc[rfm['F_score'] >= 3, 'Segment'] = 'Loyal'

# Potential: Recent buyers, not frequent
rfm.loc[rfm['R_score'] >= 3, 'Segment'] = 'Potential'

# At Risk: Low recency
rfm.loc[rfm['R_score'] == 2, 'Segment'] = 'At Risk'

# Check first few results
rfm[['RFM_Score', 'R_score', 'F_score', 'Segment']].head()
```
